```python
import pandas as pd 
import scipy 
import numpy as np
from scipy.stats import entropy
import statistics
import json

df = pd.read_excel('/home/ansost/repos/CaSpaSyn/data/pscb1.xlsx', sheet_name=1) 
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df
```

```python
# Make verb list and remove nan values.
verbs = df['Verb'].to_list()
del verbs[0:6]
del verbs[-1]
verbs = set(verbs)

ratings = {}
for verb in verbs:
    ratings[verb] = {'pronominal': [], 'active': []}
    for index, line in enumerate(df['Verb']):
        if index > 5:
            if line == verb:
                reihe = df.loc[index, :].values.tolist()[1:15]
                if 'X' in reihe or 'NA' in reihe:
                    reihe = list(df.loc[index, ][1:15].replace('X', int(0)))
                    
                if df.at[index, 'condition '] == 'pronominal': 
                    ratings[verb]['pronominal'] = ratings[verb]['pronominal'] + reihe
                elif df.at[index, 'condition '] == 'active':
                    ratings[verb]['active'] = ratings[verb]['active'] + reihe
```

```python
verb_means = {}
for verb in verbs:
    verb_means[verb] = {'pronominal': [], 'active': []}
    verb_means[verb]['pronominal'] = round(statistics.mean(ratings[verb]['pronominal']),2)
    verb_means[verb]['active'] = round(statistics.mean(ratings[verb]['active']),2)
verb_means
with open('data/pscb1-verb-means.json', 'w', encoding='utf-8') as f:
    json.dump(verb_means, f, ensure_ascii=False, indent=4)
```
